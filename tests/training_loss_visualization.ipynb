```python
# Imports
import matplotlib.pyplot as plt
import numpy as np
from daikon.models import TrainingResults 
```

```python
# Constants
training_results = "../workflows/results/brain-tumor-dataset/training/vgg16_pretrained.pkl"
```

```python
results = TrainingResults.load(training_results)
```

```python
x_labels = np.arange(0, results.loss.shape[0])

plt.plot(x_labels, results.loss)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.suptitle(f"{results.model_name} Training Loss")
plt.title(f"(Epochs: {results.loss.shape[0]}; learning rate: {results.learning_rate})", fontsize=10)
```

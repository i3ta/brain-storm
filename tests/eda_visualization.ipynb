```python
import os

import numpy as np
import matplotlib.pyplot as plt

from daikon.eda import DataAnalysisOutput
```

```python
# Variables
file = "brain-tumor-dataset-50/summary/pituitary_data_summary.pkl"

# Constants
base_dir = "../workflows/results/"
```

```python
file = os.path.join(base_dir, file)
data = DataAnalysisOutput.load(file)
```

```python
dims = data.dims[:, :2]

dim_values, dim_counts = np.unique(dims, axis=0, return_counts=True)

labels = [f"{w}x{h}" for w, h in dim_values]

plt.figure(figsize=(10, 5))
plt.bar(labels, dim_counts)
plt.xlabel("Image Dimensions")
plt.ylabel("Number of Images")
plt.title("Image Counts by Dimension")
plt.show()
```
